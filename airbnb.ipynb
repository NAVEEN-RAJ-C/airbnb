```python
# importing required libraries
from pymongo.mongo_client import MongoClient
import urllib.parse
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# MongoDB Atlas connection
username = "naveenraj"
password = "12345"
database_name = "sample_airbnb"
collection_name = "listingsAndReviews"

# Escape the username and password
escaped_username = urllib.parse.quote_plus(username)
escaped_password = urllib.parse.quote_plus(password)

# Create the MongoDB connection URI
mongo_connection = f"mongodb+srv://{escaped_username}:{escaped_password}@dsprojects.bmtisax.mongodb.net/?retryWrites=true&w=majority"

# Connect to the MongoDB cluster
client = MongoClient(mongo_connection)
mydb = client[database_name]
mycol = mydb[collection_name]
```

```python
# Retrieve all documents from the collection
all_documents = mycol.find()
doc_list = []
# Loop through and add each document to list
for document in all_documents:
    doc_list.append(document)
# create pandas dataframe
df = pd.DataFrame(doc_list)
df
```

```python
# find the number of null values
df.isnull().sum()
```

```python
# calculate the mode of first review column
first_review_mode = df['first_review'].mode()[0]
# fill NaN with the calculated mode
df['first_review'].fillna(first_review_mode, inplace = True)
```

```python
# calculate the mode of last_review column
last_review_mode = df['last_review'].mode()[0]
# fill NaN with the calculated mode
df['last_review'].fillna(last_review_mode, inplace = True)
```

```python
# calculate the mode of bedrooms column
bedrooms_mode = df['bedrooms'].mode()[0]
# fill NaN with the calculated mode
df['bedrooms'].fillna(bedrooms_mode, inplace = True)
```

```python
# calculate the mode of beds column
beds_mode = df['beds'].mode()[0]
# fill NaN with the calculated mode
df['beds'].fillna(beds_mode, inplace = True)
```

```python
# calculate the mode of bathrooms column
bathrooms_mode = df['bathrooms'].mode()[0]
# fill NaN with the calculated mode
df['bathrooms'] = df['bathrooms'].fillna(bathrooms_mode)
```

```python
# Convert 'security_deposit' column to float for mean calculation
df['security_deposit'] = df['security_deposit'].apply(lambda x: float(str(x)))
# fill NaN of security_deposit mean
df['security_deposit'] = df['security_deposit'].fillna(df['security_deposit'].mean())
```

```python
# Convert 'cleaning_fee' column to float for mean calculation
df['cleaning_fee'] = df['cleaning_fee'].apply(lambda x: float(str(x)))
# fill NaN of cleaning_fee mean
df['cleaning_fee'].fillna(df['cleaning_fee'].mean(), inplace = True)
```

```python
# Convert 'weekly_price' column to float for mean calculation
df['weekly_price'] = df['weekly_price'].apply(lambda x: float(str(x)))
# fill NaN of weekly_price mean
df['weekly_price'].fillna(df['weekly_price'].mean(), inplace = True)
```

```python
# Convert 'monthly_price' column to float for mean calculation
df['monthly_price'] = df['monthly_price'].apply(lambda x: float(str(x)))
# fill NaN of monthly_price mean
df['monthly_price'].fillna(df['monthly_price'].mean(), inplace = True)
```

```python
# fill NaN of reviews_per_month with 0
df['reviews_per_month'].fillna(0, inplace = True)
```

```python
# find the number of null values
df.isnull().sum()
```

```python
# add coordinates column to the dataframe using lambda function
df['coordinates'] = df['address'].apply(lambda x: x['location']['coordinates'])
df['country'] = df['address'].apply(lambda x: x['country'])
df
```

```python
# Save the DataFrame as a CSV file
df.to_csv('airbnb.csv', index=False)
```

```python
# Convert Decimal128 to float64 for visualization
df['price'] = df['price'].apply(lambda x: float(str(x)))
```

```python
# Set Seaborn style
sns.set(style="whitegrid")

# Create a bar plot using Seaborn
plt.figure(figsize=(15, 6))
sns.barplot(x='property_type', y='price', data=df)
plt.xlabel('property_type')
plt.ylabel('price')
plt.title('Price Distribution')
plt.xticks(rotation=90)
plt.show()
```
